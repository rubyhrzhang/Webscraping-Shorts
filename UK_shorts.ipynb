This is a simple case where we can just scrape and download the data from a specific URL (as it just opens directly as a downloaded file)

```python
import requests

import os
import pandas as pd
from datetime import datetime
```

```python
# send a GET request to the ULR
target_url = 'https://www.fca.org.uk/publication/data/short-positions-daily-update.xlsx'

access_time = datetime.now().strftime("%Y-%m-%d_%H%M%S")
response = requests.get(target_url)
response.raise_for_status() # Check for HTTP errors
```

```python
# read response as pandas df
data = pd.read_excel(response.content) # response.content is a .xlsx file
data
```

```python
# save in desired directory
working_directory = os.getcwd()
downloads_dir = os.path.join(working_directory, "downloads")

export_path = os.path.join(downloads_dir, f'UK_shorts_{access_time}.csv')
data.to_csv(export_path, index=False)  # index=False avoids writing the row index to the file
```
